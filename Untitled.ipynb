```python
def convert_to_date(dataframe: pd.DataFrame, date_column: str) -> pd.DataFrame:
    """
    Convert a string date column to datetime and extract only the date part.
    
    Args:
        dataframe (pd.DataFrame): The input DataFrame containing the date column
        date_column (str): The name of the column to convert
        
    Returns:
        pd.DataFrame: DataFrame with the converted date column
    """
    # First convert string to datetime objects with time information
    dataframe[date_column] = pd.to_datetime(
        dataframe[date_column], 
        format='%m/%d/%Y %H:%M', 
        errors='coerce'
    )
    
    # Then extract only the date part (removing time information)
    dataframe[date_column] = dataframe[date_column].dt.date
    
    return dataframe

# Apply the conversion to the ORDERDATE column
df = convert_to_date(df, 'ORDERDATE')
```

```python
def extract_date_only(dataframe: pd.DataFrame, date_column: str) -> None:
   
  
    dataframe[date_column] = pd.to_datetime(dataframe[date_column], errors='coerce').dt.date


extract_date_only(df, 'ORDERDATE')
```
