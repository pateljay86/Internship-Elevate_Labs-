```python
data = pd.read_csv(r"C:\Users\admin\Downloads\titanic\test.csv")
```

```python
# Drop high-missing column
data.drop(columns=['Cabin', 'Name', 'Ticket'], inplace=True)

# Fill missing Age and Fare
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Fare'].fillna(data['Fare'].median(), inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')

# Histogram: Age distribution
plt.figure(figsize=(8, 4))
sns.histplot(data['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

```

```python
# Countplot: Sex distribution
sns.countplot(x='Sex', data=data)
plt.title("Passenger Gender Distribution")
plt.show()


```

```python
# Countplot: Pclass
sns.countplot(x='Pclass', data=data)
plt.title("Passenger Class Distribution")
plt.show()

```

```python
# Boxplot: Age vs Pclass
sns.boxplot(x='Pclass', y='Age', data=data)
plt.title("Age by Passenger Class")
plt.show()

```

```python
# Heatmap: Correlation
plt.figure(figsize=(8, 5))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```
